**Syntax**

```python
data['Zip'].fillna(data['Zip'].mode()[0], inplace=True)
```

```python
data['Is Fraud?'] = data['Is Fraud?'].map({'No': 0, 'Yes': 1})
```

```python
data.head()
```

```python
data=data.drop('Errors?',axis=1)
```

```python
data.head()
```

```python
# Convert Amount column to numeric
data['Amount'] = data['Amount'].replace('[\$,]', '', regex=True).astype(float)
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = [ 'Use Chip', 'Merchant City', 'Merchant State']
encoder = LabelEncoder()
for col in categorical_cols:
    data[col] = encoder.fit_transform(data[col])
```

```python
from sklearn.feature_selection import mutual_info_regression
X = data.drop(['Is Fraud?', 'Time'], axis=1)
y=data['Is Fraud?']
mi_scores = mutual_info_regression(X, y)
```

```python
mi_scores_df = pd.DataFrame({
    'Feature': X.columns,
    'Mutual Information': mi_scores
}).sort_values(by='Mutual Information', ascending=False)

print(mi_scores_df)
```
