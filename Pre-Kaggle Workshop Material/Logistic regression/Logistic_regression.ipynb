```python
#Reading the csv file
data=pd.read_csv('/content/drive/MyDrive/diabetes.zip')
data.head()
```

```python
data.columns = data.columns.str.strip()
```

```python
print(data['Glucose'])
```

```python
#Splitting training and testing data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(data[['Glucose']],data.Outcome,test_size=0.3,random_state=27)
```

```python
#Training the model
from sklearn.linear_model import LogisticRegression
regModel=LogisticRegression()
regModel.fit(xtrain,ytrain)
```

```python
#Testing the model
ypred=regModel.predict(xtest)
print(ypred)
```

```python
#Testing the model(Method2)
regModel.predict_proba(xtest)
```

```python
#Model Evaluation
regModel.score(xtest,ytest)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(ytest,ypred)
print(confusion_matrix)
```

```python
#Report
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```
